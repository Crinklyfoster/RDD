Create dataset (text file) with fields like ‘Student Name’, ‘Institute’, ‘Program Name’, and
‘Gender’ and solve following questions.
1. Compute number of students from each Institute.
2. Number of students enrolled to any program.
3. Number of ‘boy’ and ‘girl’ students.
4. Number of ‘boy’ and ‘girl’ students from selected Institute.


```python
import pyspark
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName('RDD').getOrCreate()

```

```python
textRDD=spark.sparkContext.textFile('/content/drive/My Drive/Colab Notebooks/RDD Assignment/StudentDetails.txt')
textRDD.collect()
```

```python
dataRdd=textRDD.map(lambda x:x.split(' '))
dataRdd.collect()
```

```python
#1. Compute number of students from each Institute.

institue_count=dataRdd.map(lambda x:(x[1],1)).reduceByKey(lambda x,y:x+y)
institue_count.collect()
```

```python
## 2. Compute the number of students enrolled in any program
student_count=dataRdd.map(lambda x:(x[2],1)).reduceByKey(lambda x,y:x+y)
student_count.collect()
```

```python
#Number of ‘boy’ and ‘girl’ students.

gender_count=dataRdd.map(lambda x:(x[3],1)).reduceByKey(lambda x,y:x+y)
gender_count.collect()
```

```python
#4. Number of ‘boy’ and ‘girl’ students from selected Institute.

institue_gender_count=dataRdd.map(lambda x:((x[1],x[3]),1)).reduceByKey(lambda x,y:x+y)
institue_gender_count.collect()
```
