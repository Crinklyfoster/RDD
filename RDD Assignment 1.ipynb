```python
#9. Find the longest length of word from given set of words.

parRDD=spark.sparkContext.parallelize(['ABC','Class','is','very','boring','noice','JavaPython','is'])
word_lengths_rdd = parRDD.map(lambda word: (word, len(word)))
print(word_lengths_rdd .collect())

longest_word=word_lengths_rdd.reduce(lambda x,y:x if x[1] >y[1] else y)

print(f"The longest word is '{longest_word[0]}' with length {longest_word[1]}.")
```

```python
#10.Map the Registration numbers to corresponding branch. 58000 series BDA, 57000 series AIML, 38000 series VLSI, 39000 series ES, and 47000 series CDC.
#Given registration number, generate a key-value pair of Registration Number and Corresponding Branch.

register_number_RDD=spark.sparkContext.parallelize(['58001', '57032', '38017', '39012', '47005', '58020', '57008'])
register_number_RDD.collect()

def get_branch(register_number_RDD):
  if 58000<=int(register_number_RDD)<=58900:
    return 'BDA'
  elif 57000<=int(register_number_RDD)<=57900:
    return 'AIML'
  elif 38000<=int(register_number_RDD)<=38900:
    return 'VLSI'
  elif 39000<=int(register_number_RDD)<=39900:
    return 'ES'
  elif 47000<=int(register_number_RDD)<=47900:
    return 'CDC'
  else:
    return 'Unknown Branch'

branch_RDD=register_number_RDD.map(lambda x:(x,get_branch(x)))


for reg, branch in branch_RDD.collect():
    print(f"Registration Number: {reg} -> Branch: {branch}")
```

```python
#11.Text file contain numbers. Numbers are separated by one white space. There is no order to store the numbers. One line may contain one or more numbers.
#Find the maximum, minimum, sum and mean of numbers.

nums_RDD = spark.sparkContext.textFile('/content/drive/My Drive/Colab Notebooks/RDD Assignment/nums.txt')
print(nums_RDD.collect())

def cleanNums(data):
    pat = ['1', '2', '3', '4', '5', '6', '7', '8', '9', '0', '.']
    newNum = ''
    for ele in data:
        if ele in pat:
            newNum += ele
    return newNum

maxRdd=nums_RDD.map(lambda d:cleanNums(d)).filter(lambda d:len(d)>0).map((lambda d:float(d)))
print(maxRdd.collect())

print(f"The minium value is: {maxRdd.min()}")
print(f"The maximum value is:{maxRdd.max()}")
print(f"The sum of values is: {maxRdd.sum()}")
print(f"The mean of values is: {maxRdd.mean()}")
```

```python
#12.A text file (citizen.txt) contains data about citizens of country. Fields (information in file) are Name, dob, Phone, email and state name. Another file
#contains mapping of state names to state code like Karnataka is codes as KA, TamilNadu as TN, Kerala KL etc. Compress the citizen.txt file by
#changing full state name to state code.



citizenRdd=spark.sparkContext.textFile('/content/drive/My Drive/Colab Notebooks/RDD Assignment/citizen.txt')
citizenRdd.collect()

stateRdd=spark.sparkContext.textFile('/content/drive/My Drive/Colab Notebooks/RDD Assignment/state.txt')
stateRdd.collect()

state_dict = stateRdd.map(lambda x: x.split(',')).map(lambda x: (x[0].strip(), x[1].strip())).collectAsMap()

state_dict_broadcast = spark.sparkContext.broadcast(state_dict)

# Print the broadcasted value
print(state_dict_broadcast.value)

def replace_state(line):
    fields = line.split(',')  # Split by comma to get individual fields
    state_name = fields[-1].strip()  # State is the last field
    state_code = state_dict_broadcast.value.get(state_name, state_name)  # Get code or keep name if not found
    fields[-1] = state_code  # Replace the state name with the code
    return ', '.join(fields)  # Return the updated line


compressed_citizen_rdd = citizenRdd.map(replace_state)
print(compressed_citizen_rdd.collect())

#compressed_citizen_rdd.saveAsTextFile("/content/drive/My Drive/Colab Notebooks/RDD Assignment/compressed_citizen.txt")


```
